```python
import pandas as pd
```

```python
df = pd.read_spss("./KEHR81FL.SAV")
```

```python
df.head()
```

```python
# Assuming df is your DataFrame
# Define your naming format using a regex pattern
column_format_pattern = r'^SML16A\$[0-9]{2}$'

# Use filter to select columns based on the naming format
selected_columns = df.filter(regex=column_format_pattern)

# Create a new column 'most_common_result' with the most common value across the selected columns

# df[selected_columns.columns]=df[selected_columns.columns].astype("object")

df['most_common_result'] = df[selected_columns.columns].mean(axis=1)
# Display the updated DataFrame
print(df['most_common_result'])
```
