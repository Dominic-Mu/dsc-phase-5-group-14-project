```python
# Calculate the threshold for 80% empty columns
threshold = 0.8 * len(df)

# Drop columns with 80% or more empty values
df_cleaned = df.dropna(axis=1, thresh=threshold)

# Display the cleaned DataFrame
df_cleaned
```

```python
df_cleaned = df_cleaned.copy()
df_cleaned.replace('', np.NaN, inplace=True)
```

```python
df_cleaned
```

```python
df_cleaned.isna().sum()
```

```python
# Assuming df_cleaned is your DataFrame
nan_info = df_cleaned.isna().sum()
nan_percentage = (nan_info / len(df_cleaned)) * 100

# Create a DataFrame with columns and their NaN percentages
nan_df = pd.DataFrame({'Column': nan_info.index, 'NaN Count': nan_info.values, 'NaN Percentage': nan_percentage.values})

# Filter columns with NaN values
columns_with_nan = nan_df[nan_df['NaN Count'] > 0]

# Display the columns with NaN values and their percentages
columns_with_nan
```

```python
# Assuming columns_with_nan is your DataFrame with NaN information
columns_to_drop = columns_with_nan[columns_with_nan['NaN Percentage'] == 100]['Column']

# Drop columns with 100% NaN values
df_cleaned_dropped = df_cleaned.drop(columns=columns_to_drop)

# Display the DataFrame after dropping columns with 100% NaN values
df_cleaned_dropped
```

```python
# Exempt 'HV000' and 'HHID' column from the conversion
columns_to_convert = df_cleaned_dropped.columns.difference(['HV000', 'HHID'])

# Convert non-numeric values to NaN
df_cleaned_dropped_numeric = df_cleaned_dropped[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Calculate the mean for each column
column_means = df_cleaned_dropped_numeric.mean()

# Fill NaN values with the mean for each column
df_cleaned_filled = df_cleaned_dropped_numeric.fillna(column_means)

# Display the DataFrame after filling NaN values with the mean
df_cleaned_filled
```

```python
# Select and concatenate the original columns 'HV000' and 'HHID'
df_cleaned_filled = pd.concat([df_cleaned_dropped[['HHID', 'HV000']], df_cleaned_filled], axis=1)

# Display the final DataFrame after filling NaN values with the mean and arranging columns
df_cleaned_filled
```
